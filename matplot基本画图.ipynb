#### 添加文本

```python
##添加文本
plt.plot(x,y,linestyle='-',marker="o",alpha=0.1)
for i,j in zip(x,y):
    plt.text(i-0.5,j+0.03,"%s"%round(j,1),fontsize=10,color='r',
             #bbox=dict(color='red', alpha=0.3)
            )
# plt.text(x[-1],y[-1],"%s"%'X',fontsize=10,color='g') 
plt.annotate("%s"%'X',(x[-1],y[-1]),fontsize=10,color='g')#注意和text区别
plt.show()
```

#### 添加标题

```python
##添加标题
font_name = "simhei"
mpl.rcParams[u'font.sans-serif'] = font_name
mpl.rcParams['axes.unicode_minus'] = False # in case minus sign is shown as box

plt.plot(x,y,linestyle='-',marker="o",alpha=0.1)
for i,j in zip(x,y):
    plt.text(i-0.5,j+0.03,"%s"%round(j,1),fontsize=10,color='r',
             #bbox=dict(color='red', alpha=0.3)
            )

# plt.title(label="随机曲线",loc='center',color='r',fontsize=24,pad=10) 
plt.title(label="随机曲线",loc='center',
          fontdict=dict(color='r',fontsize=24)
          ,pad=10) 


"""
{'fontsize': rcParams['axes.titlesize'],
 'fontweight' : rcParams['axes.titleweight'],
 'color' : rcParams['axes.titlecolor'],
 'verticalalignment': 'baseline',
 'horizontalalignment': loc}
"""
plt.show()
```

```python
# 散点图
#cmap颜色范围参考：https://matplotlib.org/users/colormaps.html 
#plt.get_cmap('RdYlBu')
##c=y 颜色值，vmin=0,vmax=100（可选） 颜色值最大最小
```

```python
cmap = plt.get_cmap('RdYlBu')#c=y 颜色值，vmin=0,vmax=100（可选） 颜色值最大最小
plt.scatter(x,y,c=y,vmin=0,vmax=100,cmap=cmap)
plt.show()
```

#### scatter增加渐变色条

```python
cmap = plt.get_cmap('RdYlBu')#c=y 颜色值，vmin=0,vmax=100（可选） 颜色值最大最小
plt.scatter(x,y,c=y,cmap=cmap)
cbar = plt.colorbar(orientation='horizontal',
                    extend='both',
#                     fraction=0.4
                    label='单位：℃',
                    format="%.1f",#刻度小数点
                    ticks=np.linspace(0,y.max(),10),#设置刻度
                    drawedges=True,#色条内部是否有分水线
                   )

#extend='both' 两头都有值
#orientation ：cbar取向，vertical：默认垂直 horizontal：水平

plt.show()
```
